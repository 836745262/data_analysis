```python
transactions = pd.read_csv('Prod.csv',encoding='gbk')
transactions.head()
```

```python
data_filter = ~transactions.PROD.isin(['本地通话', '来电提醒','短信回执'])
transactions_filter = transactions[data_filter]

basket = transactions_filter['PROD'].groupby(transactions_filter.ID).apply(list)
basket.head()
```

```python
basket_list = basket.tolist()
rules = apriori.arules(basket_list,min_support=0.02)
rules.head()
```

```python
conf = rules[(rules.confidence>0.1)&(rules.lift>1)]
conf.sort_values(by='support',ascending=True)
```

```python
mapping = transactions_filter.PROD.drop_duplicates().to_dict() # 生成字典，(编号:产品)对
inverse_mapping = {k:v for v,k in mapping.items()} # 生成字典，(产品:编号)对
transactions_filter['PROD'] = transactions_filter['PROD'].replace(inverse_mapping)
transactions_filter.head()
```

```python
basket = transactions_filter.PROD.groupby(transactions_filter.ID).apply(list)
basket.head()
```

```python
basket = basket.tolist()

from orangecontrib.associate.fpgrowth import *

n_examples = len(transactions_filter)
```

```python
itemsets = dict(frequent_itemsets(basket,0.02))
ass_rules = association_rules(itemsets,min_confidence=.1)
stas = rules_stats(ass_rules,itemsets,n_examples)
```

```python
result = pd.DataFrame(stas,columns=['atecedent', 'consequent', 'support', 
                                      'confidence', 'coverage', 'strength', 
                                      'lift', 'leverage'])
result.head()
```

用产品名称替换产品编号

```python
pid_to_product = lambda fsets: tuple(mapping[fset] for fset in fsets)
result['atecedent'] = result['atecedent'].map(pid_to_product)
result['consequent'] = result['consequent'].map(pid_to_product)
result.head()
```
